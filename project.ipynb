```python
import pandas as pd

file_path = 'data/SYRCityline_Requests_(2021-Present) (1).csv'
df = pd.read_csv(file_path)

# Convert the 'Created_at_local' column to datetime if it's not already in datetime format
df['Created_at_local'] = pd.to_datetime(df['Created_at_local'])

# Filter rows where the Category is 'Potholes'
potholes_df = df[df['Category'] == 'Potholes']

# Group by 'Created_at_local' and count the number of rows for each date
potholes_count_by_date = potholes_df.groupby(potholes_df['Created_at_local'].dt.date).size()

print(potholes_count_by_date)
```

```python
# Convert the index of potholes_count_by_date to datetime and rename it to 'Date'
potholes_count_by_date = potholes_count_by_date.reset_index()
potholes_count_by_date.columns = ['Date', 'Potholes_Count']
potholes_count_by_date['Date'] = pd.to_datetime(potholes_count_by_date['Date'])

# Display the modified DataFrame
potholes_count_by_date
```

```python
potholes_count_by_date.info()
```

```python
file_path = 'data/40_Year_Temperature_Dataset_SU.csv'
weather_df = pd.read_csv(file_path, skiprows=7)

# Convert 'Year', 'Month', and 'Day' columns to a single datetime column
weather_df['Date'] = pd.to_datetime(weather_df[['Year', 'Month', 'Day']])
```

```python
# Filter rows from 2021 onwards
weather_df = weather_df[weather_df['Year'] >= 2021]

weather_df
```

```python
weather_df.info()
```

```python
# Merge potholes_count_by_date as a new column
merged_df = pd.merge(weather_df, potholes_count_by_date, on='Date', how='left')

# Rename the merged column to 'Potholes_Count'
merged_df.rename(columns={0: 'Potholes_Count'}, inplace=True)

# Display the modified DataFrame
merged_df
```

### Date vs Temp at Recording

```python
# @title Date vs Temp at Recording

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date']
  ys = series['Temp at Recording']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = merged_df.sort_values('Date', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date')
_ = plt.ylabel('Temp at Recording')
```

### Temp at Recording

```python
from matplotlib import pyplot as plt
merged_df['Temp at Recording'].plot(kind='hist', bins=20, title='Temp at Recording')
plt.gca().spines[['top', 'right',]]
```

### Temperature vs Humidity Over Time

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5.2))
plt.scatter(merged_df['Date'], merged_df['Temp at Recording'], s=merged_df['Potholes_Count'], c=merged_df['Relative Humidity'], cmap='viridis')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
_ = plt.colorbar(label='Relative Humidity (%)')
```

### Handling null values

```python
merged_df = merged_df.dropna()
merged_df
```

### Correlation Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = merged_df.corr()

# Set up the heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", ax=ax, cmap="coolwarm")

# Add a title and labels
plt.title("Correlation Matrix of Weather and Potholes Data")
plt.xlabel("Variables")
plt.ylabel("Variables")

# Show the heatmap
plt.show()
```
